```python
from ultralytics import YOLO
import os

model = YOLO("yolov8n-seg.pt")

input_dir = "/content/video_frames"
output_dir = "/content/segmented_results"
os.makedirs(output_dir, exist_ok=True)

frame_files = sorted([f for f in os.listdir(input_dir) if f.endswith(".jpg")])

for i, file in enumerate(frame_files):
    img_path = os.path.join(input_dir, file)
    result = model(img_path)


    result[0].save(filename=os.path.join(output_dir, f"frame_{i:03d}.jpg"))

print(f" Saved {len(frame_files)} segmented images to: {output_dir}")
```
